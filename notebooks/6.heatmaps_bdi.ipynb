```python
%matplotlib notebook
import sys
sys.path.append('/Users/priya/Desktop/heatsequer')
import heatsequer as hs
import csv
import numpy as np
```

```python
mm=hs.load('../data/otu_100nt_235samples.biom', '../data/ag235_w_scores_alpha.txt')
```

```python
#sort by age
mm=hs.sortsamples(mm,'bin_bdi21_score',numeric=True)
```

```python
mmc=hs.clusterbacteria(mm,10)
```

```python
#getting differentially abundant bacteria
dd=hs.getdiffsig(mm,'bin_bdi21_score','1','4','ranksum')
```

```python
dd=hs.getpfdr(mm,'bin_bdi21_score','1','4','ranksum',maxfval=0.25)
```

```python
ddc=hs.clusterbacteria(dd, 10)
ddc=hs.filtersamples(ddc, 'bin_bdi21_score', ['1','4'])
hs.plotexp(ddc,'bin_bdi21_score')
```
